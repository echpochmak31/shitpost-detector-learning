```python
import pandas as pd
import random
```

```python
shitpost_df = pd.read_csv('./data/filtered_shitpost_df.csv', index_col=None)
shitpost_df
```

```python
wiki_df = pd.read_csv('./data/filtered_wiki_df.csv', index_col=None)
wiki_df
```

```python
result = pd.concat([shitpost_df, wiki_df])
result
```

```python
shuffled_df = result.sample(frac=1, random_state=random.seed()) 

# Сброс индексов и построение их заново от 0
shuffled_df.reset_index(drop=True, inplace=True)
shuffled_df
```

```python
shuffled_df.to_csv('./data/shitpost_classification_df.csv', index=False)
```

```python
filtered_df = shuffled_df[shuffled_df['is_shitpost'] == False]
filtered_df
```
