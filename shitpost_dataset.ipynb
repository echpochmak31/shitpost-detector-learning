```python
import neattext as nt
import re
import pandas as pd
```

```python
def preprocess_text(text):
    text = re.sub(r'[^\u0400-\u04FF\s.,;?!]', '', text)

    docx = nt.TextFrame(text)
    docx.remove_puncts()
    docx.remove_emojis()
    docx.remove_phone_numbers()
    docx.remove_numbers()
    docx.remove_emails()
    docx.remove_stopwords(lang='ru')
    docx.normalize(level='deep')

    text = docx.text.strip()

    return text

def preprocess_file(filepath):
    processed_lines = []
    ignore_patterns = re.compile(r'украи|росс|росси|украин', re.IGNORECASE)
    
    with open(filepath, 'r', encoding='utf-8') as f:
        for line in f:
            if not ignore_patterns.search(line):  # Проверяем, не содержит ли строка запрещенные паттерны
                processed_line = preprocess_text(line)
                if processed_line:  # Проверяем, не пустая ли строка
                    processed_lines.append(processed_line)
                
    return processed_lines
```

```python

filepath = './data/shitpost/misha_shitpost.txt'

processed_lines = preprocess_file(filepath)
for line in processed_lines:
    print(line)
print(len(processed_lines))
```

```python
import os

directory_path = '.\data\shitpost'

all_shitpost_line = []
for filename in os.listdir(directory_path):
    filepath = os.path.join(directory_path, filename)

    # Проверяем, является ли объект по указанному пути файлом
    if os.path.isfile(filepath):
        print(filepath)
        processed_lines = preprocess_file(filepath)
        all_shitpost_line += processed_lines
```

```python
shitpost_df = pd.DataFrame(all_shitpost_line, columns=['text'])

shitpost_df['is_shitpost'] = True

shitpost_df
```
